```python
simulator = QasmSimulator()
params= 3/4*pi
def vqe_ansatz_1(params):
    q = QuantumRegister(4)
    c = ClassicalRegister(4)
    ansatz = QuantumCircuit(q,c)
    #GHZ-like state
    ansatz.h(0)
    ansatz.cx(0,1)
    ansatz.cx(1,2)
    ansatz.cx(2,3)
    ansatz.rz(params,q[0])
    return ansatz
```
