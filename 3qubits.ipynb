```python
#使用的simulator
simulator = QasmSimulator()
#GHZ-states得知之最大角度帶入
params= 2/4*pi
#定義GHZ的quantum circuit
def vqe_ansatz_1(params):
    q = QuantumRegister(3)
    c = ClassicalRegister(3)
    ansatz = QuantumCircuit(q,c)
    #GHZ-like state
    ansatz.h(0)
    ansatz.cx(0,1)
    ansatz.cx(1,2)
    ansatz.rz(params,q[0])
    return ansatz
```
